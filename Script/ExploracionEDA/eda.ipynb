```python
# notebooks/eda_simulacion_estudiantes.ipynb

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configuraciones generales
sns.set(style="whitegrid")

# Cargar datos
ruta_csv = '../simulacion_estudiantes.csv'
df = pd.read_csv(ruta_csv)
```

```python
# -----------------------------
# 1. Distribución de la variable objetivo
# -----------------------------

plt.figure(figsize=(8, 4))
sns.histplot(df['calificacion_matematica'], bins=20, kde=True)
plt.title("Distribución de Calificación en Matemática")
plt.xlabel("Calificación")
plt.ylabel("Frecuencia")
plt.show()

print("Estadísticas básicas de 'calificacion_matematica':")
print(df['calificacion_matematica'].describe())
```

```python
# -----------------------------
# 2. Correlaciones con variables numéricas
# -----------------------------

corrs = df.corr(numeric_only=True)['calificacion_matematica'].sort_values(ascending=False)
print("Correlaciones con 'calificacion_matematica':")
print(corrs)
```

```python
# -----------------------------
# 3. Visualización de las 3 variables numéricas más correlacionadas
# -----------------------------
top_vars = corrs.drop('calificacion_matematica').abs().sort_values(ascending=False).head(3).index
for var in top_vars:
    plt.figure(figsize=(6, 4))
    sns.scatterplot(data=df, x=var, y='calificacion_matematica')
    plt.title(f"Calificación en Matemática vs {var}")
    plt.show()

    correlacion = df[[var, 'calificacion_matematica']].corr().iloc[0, 1]
    tendencia = "positiva" if correlacion > 0 else "negativa"
    print(f"\nLa variable '{var}' muestra una correlación {tendencia} de aproximadamente {correlacion:.2f} con la calificación en matemática.")
    print("Esto sugiere que a medida que '{0}' {1}, la calificación tiende a {2}.".format(
        var,
        "aumenta" if correlacion > 0 else "disminuye",
        "aumentar" if correlacion > 0 else "disminuir"
    ))
```

```python
# -----------------------------
# 4. Análisis de variables categóricas
# -----------------------------
cat_cols = [
    'genero', 'seccion', 'asistencia_inicial', 'libros_en_casa', 'internet_en_casa',
    'adecuaciones_curriculares', 'tipo_neae', 'violencia_familiar',
    'enfermedad_grave_familiar', 'catastrofe_familiar', 'conducta_riesgo_observada'
]

for col in cat_cols:
    if col in df.columns:
        plt.figure(figsize=(6, 3))
        sns.boxplot(data=df, x=col, y='calificacion_matematica')
        plt.title(f"Calificación en Matemática por {col}")
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()

        resumen = df.groupby(col)['calificacion_matematica'].describe()
        print(f"\nResumen estadístico de 'calificacion_matematica' por '{col}':")
        print(resumen)
        print(f"\nLa variable categórica '{col}' parece mostrar diferencias en las calificaciones promedio. Esto podría indicar que tiene cierto poder explicativo sobre el rendimiento en matemática, dependiendo del grupo.")
```
